```python
import matplotlib.pyplot as plt
```

## Plot top-1 val/training accuracy as a function of number of layers frozen:

```python
acc_val = [0.4609, 0.5109, 0.5308, 0.6117, 0.6620, 0.6683, 0.6637, 0.6670, 0.6662, 0.6701, 0.6662, 0.6694, 0.6632, 0.6683]
acc_training = [0.335, 0.4102, 0.4250, 0.5885, 0.7523, 0.7932,0.7684, 0.7930, 0.7939, 0.7945, 0.7839, 0.8226, 0.8297, 0.8362]
layers = [0,1,2,4,8,16,18,20,24,28,32,36,40,44]


plt.plot(layers, acc_val, label = 'Validation accuracy')
plt.plot(layers, acc_training, label = 'Training accuracy')
plt.xlabel('Num layers unfrozen')
plt.ylabel('Accuracy')
plt.title('EfficientNet training and validation accuracy\n Per number of unfrozen layers')
plt.legend()
```
